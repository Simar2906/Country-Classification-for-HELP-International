## Importing the Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the Dataset

```python
dataset = pd.read_csv('datasets_721951_1255613_Country-data.csv')
X = dataset.iloc[:, 1:].values
```

## Determining No. of Clusters using Elbow Method

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('cluster number')
plt.ylabel('WCSS')
plt.show()
```

## Training Kmeans model on Dataset

```python
kmeans = KMeans(n_clusters=2, init = 'k-means++', random_state = 0) # 2 clusters taken
y_kmeans = kmeans.fit_predict(X)
```
