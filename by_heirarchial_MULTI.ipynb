## Importing the Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
```

## Importing the Dataset

```python
dataset = pd.read_csv('datasets_721951_1255613_Country-data.csv')
X = dataset.iloc[:, :].values
X_1 = dataset.iloc[:, [1, 7]].values #Child Mortality Rate and Life Expectancy
X_2 = dataset.iloc[:, [1, 8]].values #Child Mortality Rate and Total Fertility
X_3 = dataset.iloc[:, [2, 4]].values #Import and Export
X_4 = dataset.iloc[:, [3, 5]].values #Health and Income
X_5 = dataset.iloc[:, [3, 9]].values #Health and GDP
X_6 = dataset.iloc[:, [3, 7]].values #Health and Life Expectancy
```

## Determining No. of clusters using Dendrogram

### For X_1

```python
dendrogram = sch.dendrogram(sch.linkage(X_1, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Countries')
plt.ylabel('Euclidean Distance')
plt.show()
```

### For X_2

```python
dendrogram = sch.dendrogram(sch.linkage(X_2, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Countries')
plt.ylabel('Euclidean Distance')
plt.show()
```

### For X_4

## Training Heirarchial Model on Dataset

### For X_1

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 2, affinity='euclidean', linkage = 'ward')
y_1 = hc.fit_predict(X_1)
```

### For X_2

```python
hc = AgglomerativeClustering(n_clusters = 2, affinity='euclidean', linkage = 'ward')
y_2 = hc.fit_predict(X_2)
```

### For X_3

```python
hc = AgglomerativeClustering(n_clusters = 2, affinity='euclidean', linkage = 'ward')
y_3 = hc.fit_predict(X_3)
```

### For X_4

```python
hc = AgglomerativeClustering(n_clusters = 2, affinity='euclidean', linkage = 'ward')
y_4 = hc.fit_predict(X_4)
```

### For X_5

```python
hc = AgglomerativeClustering(n_clusters = 2, affinity='euclidean', linkage = 'ward')
y_5 = hc.fit_predict(X_5)
```

### For X_6

```python
hc = AgglomerativeClustering(n_clusters = 2, affinity='euclidean', linkage = 'ward')
y_6 = hc.fit_predict(X_6)
```
