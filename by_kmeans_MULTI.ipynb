## Importing the Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
```

## Importing the Dataset

```python
dataset = pd.read_csv('datasets_721951_1255613_Country-data.csv')
X = dataset.iloc[:, :].values
X_1 = dataset.iloc[:, [1, 7]].values #Child Mortality Rate and Life Expectancy
X_2 = dataset.iloc[:, [1, 8]].values #Child Mortality Rate and Total Fertility
X_3 = dataset.iloc[:, [2, 4]].values #Import and Export
X_4 = dataset.iloc[:, [3, 5]].values #Health and Income
X_5 = dataset.iloc[:, [3, 9]].values #Health and GDP
X_6 = dataset.iloc[:, [3, 7]].values #Health and Life Expectancy
```

## Determining No. of clusters by Elbow Method

### For X_1

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 0)
    kmeans.fit(X_1)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('cluster number')
plt.ylabel('WCSS')
plt.show()
```
